```python
Algorithm 1. Corrosion segmentation using bounding-box prompted SAM

Input:
    I            : inspection image
    B = {b_i}    : defect bounding boxes predicted by YOLO
    SAM_model    : pretrained SAM model (vit_h)

Output:
    M = {m_i}    : binary masks of corrosion regions
    C = {c_i}    : extracted 2D contours

Procedure:
1. Load image I
2. Initialize SAM_model with image encoder, prompt encoder, and mask decoder
3. Compute image embedding E_img from I
4. For each bounding box b_i in B do
5.     Encode b_i as a sparse box prompt
6.     Feed E_img and the encoded prompt into the SAM mask decoder
7.     Set multi_mask_output = False
8.     Obtain a single binary mask m_i
9.     Post-process m_i to remove small noise regions if needed
10.    Extract ordered boundary points c_i from m_i
11. End For
12. Return masks M and contours C
```
