```python
Algorithm 2. Perspective and scale correction of defect contour

Input:
    C_pix = {(x_i, y_i)}         : contour points in image pixels
    P_src = {p1, p2, p3, p4}     : four reference points in the original image
    P_dst = {q1, q2, q3, q4}     : target rectangular reference points
    mu                            : pixel pitch
    f                             : focal length
    D                             : camera-to-wall distance

Output:
    C_phy = {(X_i, Y_i)}         : contour points in physical units

Procedure:
1. Estimate homography matrix H from P_src to P_dst
2. For each contour point (x_i, y_i) in C_pix do
3.     Apply perspective correction:
           [u_i, v_i, w_i]^T = H * [x_i, y_i, 1]^T
           x'_i = u_i / w_i
           y'_i = v_i / w_i
4. End For
5. Compute scale factor:
           s = mu * D / f
6. For each corrected point (x'_i, y'_i) do
7.     Convert to physical coordinates:
           X_i = s * x'_i
           Y_i = s * y'_i
8. End For
9. Return ordered physical contour C_phy
```
