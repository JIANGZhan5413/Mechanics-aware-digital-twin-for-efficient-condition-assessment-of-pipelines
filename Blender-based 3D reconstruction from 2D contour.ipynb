```python
Algorithm 3. 3D reconstruction of corrosion geometry in Blender

Input:
    C_phy                  : ordered 2D contour points in physical units
    W_plate, H_plate       : dimensions of the unrolled pipe wall
    d_cor                  : assumed corrosion depth
    R_pipe                 : pipe radius

Output:
    S_defect               : reconstructed surface mesh of the corroded region

Procedure:
1. Import contour C_phy into Blender through the API
2. Create a rectangular planar wall with size W_plate × H_plate
3. Convert contour C_phy into a closed 2D curve object
4. Generate a defect patch from the closed curve
5. Apply Boolean difference to remove the defect shape from the planar wall
6. Extrude the defect region along the local thickness direction by depth d_cor
7. Join the defect surface and surrounding wall geometry if needed
8. Roll the planar wall to cylindrical curvature with radius R_pipe
9. Triangulate or remesh the resulting geometry if needed
10. Export the reconstructed surface mesh as STL
11. Return S_defect
```
