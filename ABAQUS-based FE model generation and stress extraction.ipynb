```python
Algorithm 4. FE model generation and stress analysis in ABAQUS

Input:
    STEP_solid                 : reconstructed corrosion geometry
    material_params            : API 5L X65 properties
    pipe_params                : pipe diameter and wall thickness
    load_cases = {p_j}         : internal pressure levels
    mesh_size_local            : local refined mesh size
    mesh_size_global           : global mesh size

Output:
    sigma_vm_max = {s_j}       : maximum von Mises stress under each load case
    u_radial = {u_j}           : radial displacement response

Procedure:
1. Import STEP_solid into ABAQUS/CAE
2. Create the pipe part and assign corrosion geometry
3. Define material properties:
       density, elastic modulus, Poisson’s ratio,
       multi-linear plastic stress–strain relation,
       von Mises yield criterion
4. Assemble the model and apply symmetry if needed
5. Apply boundary conditions at pipe ends and symmetry planes
6. Partition the defect region for local mesh refinement
7. Mesh the defect region with fine elements
8. Mesh the remaining pipe body with coarser elements
9. Enable geometric nonlinearity (Nlgeom = ON)
10. For each internal pressure p_j in load_cases do
11.     Apply internal pressure p_j
12.     Submit the ABAQUS job
13.     Check convergence; if divergence occurs, refine the mesh or adjust controls
14.     Extract maximum von Mises stress s_j
15.     Extract radial displacement u_j
16. End For
17. Return sigma_vm_max and u_radial
```
